```python
class Lodging:
    def __init__(self, date, name, category, type, rating, price, average_revenue):
        self.date = date
        self.name = name
        self.category = category
        self.type = type
        self.rating = rating
        self.price = price
        self.average_revenue = average_revenue
        self.unique_id = id(self)

    def __str__(self):
        return f"{self.unique_id}, {self.date}, {self.name}, {self.category},{self.type}, {self.rating}, {self.price}, {self.average_revenue}"

# date,name,category,rating,price,average_revenue
class Travel(Lodging):
    def __init__(self, date, name, category, rating, price, average_revenue):
        super().__init__(date, name, category, "Travel", rating, price, average_revenue)

class Vacation(Lodging):
    def __init__(self, date, name, category, rating, price, average_revenue):
        super().__init__(date, name, category, "Vacation", rating, price, average_revenue)

# date,name,rating,price,average_revenue
class HotelRoom(Travel):
    def __init__(self, date, name, rating, price, average_revenue):
        super().__init__(date, name, "HotelRoom", rating, price, average_revenue)

class Cottage(Vacation):
    def __init__(self, date, name, rating, price, average_revenue):
        super().__init__(date, name, "Cottage", rating, price, average_revenue)

class BeachHouse(Vacation):
    def __init__(self, date, name, rating, price, average_revenue):
        super().__init__(date, name, "BeachHouse", rating, price, average_revenue)



```

```python
# Ensure the 'category_encoded' column is present, which seems to be already encoded
if 'category_encoded' not in dfd.columns:
    from sklearn.preprocessing import LabelEncoder
    le = LabelEncoder()
    dfd['category_encoded'] = le.fit_transform(dfd['category'])

# Now create the user-item matrix for collaborative filtering
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Create a user-item matrix
# 'uid' (unique identifier) represents the user and 'category_encoded' is the item
item_user_matrix = dfd.pivot_table(values='rating', index=dfd.index, columns='category_encoded', fill_value=0)

# Compute the cosine similarity between items (category)
item_similarity = cosine_similarity(item_user_matrix.T)

# Predict ratings based on item similarity (Collaborative Filtering)
predicted_ratings_cf = np.dot(item_similarity, item_user_matrix.T) / np.array([np.abs(item_similarity).sum(axis=1)]).T

# Print collaborative filtering recommendations
print("Collaborative Filtering Recommendations (Item-Based):")
print(predicted_ratings_cf)

# Visualization 1: Item Similarity Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(item_similarity, cmap='coolwarm', annot=False)
plt.title('Item Similarity Heatmap (Collaborative Filtering)')
plt.xlabel('Items (Categories)')
plt.ylabel('Items (Categories)')
plt.show()

# Visualization 2: Top 5 Recommendations for a Specific User
# Select the first user (UID at index 0) for example
user_id = 0
predicted_ratings_user = predicted_ratings_cf[:, user_id]

# Bar chart showing top 5 recommended items for this user
top_5_items = np.argsort(-predicted_ratings_user)[:5]
top_5_ratings = predicted_ratings_user[top_5_items]

plt.figure(figsize=(8, 5))
sns.barplot(x=top_5_items, y=top_5_ratings, palette='viridis')
plt.title(f'Top 5 Recommended Items for User {dfd.index[user_id]}')
plt.xlabel('Item Category ID')
plt.ylabel('Predicted Rating')
plt.show()

# Visualization 3: Predicted Ratings Heatmap for All Users
plt.figure(figsize=(12, 8))
sns.heatmap(predicted_ratings_cf, cmap='YlGnBu', annot=False)
plt.title('Predicted Ratings Heatmap (Collaborative Filtering)')
plt.xlabel('Users (UID)')
plt.ylabel('Items (Categories)')
plt.show()
```

1. Item Similarity Heatmap

    What it is: This chart shows how similar different products (in this case, lodging categories like "Seaside Cottage" and "BeachHouse") are to each other based on user preferences.
    Why it matters: If two items are similar, a user who liked one of them is more likely to be recommended the other. For example, if a customer likes "Seaside Cottage," the system might also suggest "BeachHouse" if they are closely related.
    How it helps: It ensures the recommendation system suggests items that align with what users already enjoy, making it more likely theyll engage with new items.

2. Top 5 Recommendations for a User

    What it is: This chart shows the top 5 items the system recommends to a particular user based on their past behavior and preferences.
    Why it matters: It provides a personalized experience for each user by suggesting the items they are most likely to enjoy.
    How it helps: Personalization leads to better user satisfaction and increased engagement. For instance, if the system knows that User A liked certain vacation spots, it will recommend similar lodgings based on those preferences.

3. Predicted Ratings Heatmap for All Users

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity
import plotly.express as px
import plotly.io as pio


# Assume the cleaned 'dfd' dataset is already available

# Step 1: Matrix Factorization (SVD)
item_user_matrix = dfd.pivot_table(values='rating', index=dfd.index, columns='category_encoded', fill_value=0)
n_features = item_user_matrix.shape[1]  
n_components = min(5, n_features)

svd = TruncatedSVD(n_components=n_components, random_state=42)
matrix_svd = svd.fit_transform(item_user_matrix)

predicted_ratings_svd = np.dot(matrix_svd, svd.components_)

# Assuming you already computed collaborative filtering predictions earlier
# Check the shapes of the predicted matrices (CF and SVD)
print(f"Shape of predicted_ratings_cf: {predicted_ratings_cf.shape}")
print(f"Shape of predicted_ratings_svd: {predicted_ratings_svd.shape}")

# Ensure shapes are compatible
if predicted_ratings_cf.shape != predicted_ratings_svd.shape:
    if predicted_ratings_cf.shape == predicted_ratings_svd.T.shape:
        predicted_ratings_svd = predicted_ratings_svd.T

# Combine SVD predictions with Collaborative Filtering predictions
hybrid_ratings = 0.5 * predicted_ratings_cf + 0.5 * predicted_ratings_svd

# Step 2: Visualizations with Attractive Charts

import plotly.express as px
import plotly.io as pio

# Set the renderer to 'svg' to display the chart as an SVG image
pio.renderers.default = 'svg'

# Create the interactive bubble chart
fig = px.scatter(dfd, x='price', y='rating', size='average_revenue', color='category_encoded', 
                 hover_name='category', log_x=True, size_max=60)

fig.update_layout(title='Bubble Chart: Price vs Rating with Average Revenue',
                  xaxis_title='Price', yaxis_title='Rating')


# 2. Line Chart: Predicted Ratings per User (Hybrid Model)
user_ids = list(range(len(hybrid_ratings)))
avg_pred_per_user = np.mean(hybrid_ratings, axis=1)

plt.figure(figsize=(10, 6))
plt.plot(user_ids, avg_pred_per_user, marker='o', linestyle='-', color='green')
plt.title('Line Chart: Predicted Ratings per User (Hybrid Model)')
plt.xlabel('User ID')
plt.ylabel('Average Predicted Rating')
plt.show()

# 3. Area Chart: Predicted Ratings for Different Categories
category_ids = list(range(len(svd.components_)))
category_ratings = np.mean(svd.components_, axis=1)

plt.fill_between(category_ids, category_ratings, color="skyblue", alpha=0.4)
plt.plot(category_ids, category_ratings, color="Slateblue", alpha=0.6, linewidth=2)
plt.title('Area Chart: Average Predicted Ratings per Category (SVD)')
plt.xlabel('Category ID')
plt.ylabel('Predicted Rating')
plt.show()

# 4. Correlation Heatmap (For Exploratory Analysis)
plt.figure(figsize=(12, 8))
numeric_cols = dfd.select_dtypes(include=[np.number])
correlation_matrix = numeric_cols.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap: Exploring Feature Relationships')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.model_selection import train_test_split
from sklearn.linear_model import ElasticNet

# Step 1: Matrix Factorization (SVD) - already done
item_user_matrix = dfd.pivot_table(values='rating', index=dfd.index, columns='category_encoded', fill_value=0)
n_features = item_user_matrix.shape[1]  
n_components = min(5, n_features)

svd = TruncatedSVD(n_components=n_components, random_state=42)
matrix_svd = svd.fit_transform(item_user_matrix)

predicted_ratings_svd = np.dot(matrix_svd, svd.components_)

# Step 2: Ensuring compatible shapes
if predicted_ratings_cf.shape != predicted_ratings_svd.shape:
    if predicted_ratings_cf.shape == predicted_ratings_svd.T.shape:
        predicted_ratings_svd = predicted_ratings_svd.T

# Combine SVD and Collaborative Filtering predictions (hybrid model)
hybrid_ratings = 0.5 * predicted_ratings_cf + 0.5 * predicted_ratings_svd

# Step 3: Implementing Item-Based Collaborative Filtering
item_similarity = cosine_similarity(item_user_matrix.T)
predicted_ratings_cf_item = np.dot(item_user_matrix, item_similarity) / np.array([np.abs(item_similarity).sum(axis=1)])

if hybrid_ratings.shape != predicted_ratings_cf_item.shape:
    if hybrid_ratings.shape == predicted_ratings_cf_item.T.shape:
        predicted_ratings_cf_item = predicted_ratings_cf_item.T

hybrid_item_based = 0.5 * hybrid_ratings + 0.5 * predicted_ratings_cf_item

# Step 4: SLIM (Sparse Linear Method)
X_train, X_test, y_train, y_test = train_test_split(item_user_matrix, item_user_matrix, test_size=0.2, random_state=42)
slim_model = ElasticNet(alpha=0.01, l1_ratio=0.1, positive=True)
slim_model.fit(X_train, y_train)
predicted_ratings_slim = slim_model.predict(item_user_matrix)

# Step 5: Final Hybrid Model (combining SVD, CF, SLIM)
if predicted_ratings_slim.shape != hybrid_item_based.shape:
    predicted_ratings_slim = predicted_ratings_slim.T
final_hybrid_ratings = 0.3 * predicted_ratings_svd + 0.3 * predicted_ratings_cf_item + 0.4 * predicted_ratings_slim

# Step 6: Visualizations with Attractive Charts

# 1. Bubble Chart using matplotlib
plt.figure(figsize=(10, 6))
bubble_sizes = dfd['average_revenue'] / 1000  # Bubble sizes
plt.scatter(dfd['price'], dfd['rating'], s=bubble_sizes, alpha=0.5, c=dfd['category_encoded'], cmap='viridis')

plt.title('Bubble Chart: Price vs Rating with Average Revenue')
plt.xlabel('Price')
plt.ylabel('Rating')

# Show the plot with colorbar and legend
plt.colorbar(label='Category (Encoded)')
plt.show()

# 2. Line Chart: Predicted Ratings per User (Final Hybrid Model)
user_ids = list(range(len(final_hybrid_ratings)))
avg_pred_per_user = np.mean(final_hybrid_ratings, axis=1)

plt.figure(figsize=(10, 6))
plt.plot(user_ids, avg_pred_per_user, marker='o', linestyle='-', color='green')
plt.title('Line Chart: Predicted Ratings per User (Final Hybrid Model)')
plt.xlabel('User ID')
plt.ylabel('Average Predicted Rating')
plt.show()

# 3. Area Chart: Predicted Ratings for Different Categories
category_ids = list(range(len(svd.components_)))
category_ratings = np.mean(svd.components_, axis=1)

plt.fill_between(category_ids, category_ratings, color="skyblue", alpha=0.4)
plt.plot(category_ids, category_ratings, color="Slateblue", alpha=0.6, linewidth=2)
plt.title('Area Chart: Average Predicted Ratings per Category (SVD)')
plt.xlabel('Category ID')
plt.ylabel('Predicted Rating')
plt.show()

# 4. Correlation Heatmap (For Exploratory Analysis)
plt.figure(figsize=(12, 8))
numeric_cols = dfd.select_dtypes(include=[np.number])
correlation_matrix = numeric_cols.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap: Exploring Feature Relationships')
plt.show()

# Additional: Top-N Recommendations Visualization (Bar Chart)
top_n_recommendations = np.argsort(final_hybrid_ratings, axis=1)[:, -5:]  # Top 5 recommendations per user
plt.figure(figsize=(12, 6))
for i, top_items in enumerate(top_n_recommendations[:5]):  # Display top recommendations for first 5 users
    plt.bar([f'User {i+1}: Item {j}' for j in top_items], final_hybrid_ratings[i, top_items])
plt.title('Top 5 Recommendations per User')
plt.ylabel('Predicted Rating')
plt.show()
```
